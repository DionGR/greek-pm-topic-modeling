```python
import stanza
```

```python
pipeline = stanza.Pipeline(lang='el', 
                processors='tokenize,pos,lemma',
                    pos_model_path="models/octis/checkpoints/stanza/el_gdt_tagger.pt", 
                lemma_model_path="models/octis/checkpoints/stanza/el_gdt_lemmatizer.pt")
```

```python
doc = pipeline("τῷ")
```

```python
for word in doc.iter_words():
    print("word: ", word.lemma)
```

```python
import nltk
from nltk.corpus import stopwords
from spacy.lang.el.stop_words import STOP_WORDS as el_stop

gr_nltk = stopwords.words('greek')
```

```python
print(len(gr_nltk))
print(len(el_stop))
```

```python
# union of the two sets
el_stop_set = set(el_stop)

len(el_stop_set)
```

```python
el_stop_set
```

```python
union_lemmatized = set()

for word in el_stop:
    doc = pipeline(word)
    for w in doc.iter_words():
        union_lemmatized.add(w.lemma)
        
print(len(union_lemmatized))
```
